```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
import numpy as np

""" ONLY WORKS FOR 2D GRAPH """
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']
A = np.array([
    [-1,2],  # vector 1
    [-5,1],  # vector 2
    [-4,-1],  # vector 3
])


b = np.array([
    [4],
    [-16],
    [-11]
])
spacing = 0.5  # This can be your user specified spacing.
minorLocator = MultipleLocator(spacing)
x1 = np.linspace(-5, 10, 100)
plt.figure(figsize=(11, 9))

for i in range(len(A)):

    if A[i][0] != 0 and A[i][1] != 0:
        div = A[i][1]
        x2 = (x1 * A[i][0] + b[i][0])/div
        plt.plot(x1, x2, colors[i], label='({0}) {1}*x1 + {2}*x2 <= {3}'.format(
            i+1, A[i][0], div, b[i][0]), linewidth=2.0)
        #plt.fill_between(x1, x2, where=x2 <= (
        #    x1 * A[i][0] + b[i][0])/div, interpolate=True, color=colors[i], alpha=0.1)
        # continue
    # validate if is a horizontal line
    if A[i][0] == 0 and A[i][1] != 0:
        x2 = b[i][0]/A[i][1]
        plt.axhline(x2, -5, 5, c=colors[i], label='({0}) {1}*x2 <= {2}'.format(
            i+1, A[i][1], b[i][0]))
        #plt.fill_between(
        #    x1, x2, where=x2 <= b[i][0]/A[i][1], interpolate=True, color=colors[i], alpha=0.1)

    # validate if is a vertical line
    if A[i][0] != 0 and A[i][1] == 0:
        x2 = b[i][0]/A[i][0]
        data_x = np.linspace(-5, 5, 100)
        # print(data_x)
        plt.axvline(x2, -5, 5, c=colors[i], label='({0}) {1}*x1 <= {2}'.format(
            i+1, A[i][0], b[i][0]))
        #plt.fill_between(
        #    x1, x2, 0, where=x1 <= b[i][0]/A[i][0], interpolate=True, color=colors[i], alpha=0.1)

#plt.quiver([6],[5],color='y',units='width',scale=20)
plt.title('Graph')
plt.xlabel('x1', color='#1C2833')
plt.ylabel('x2', color='#1C2833')
plt.ylim(-4, 8)
plt.legend(loc='upper left')

plt.grid(visible=True, which='major', color='k', linestyle='-', alpha=0.2)
plt.grid(visible=True, which='minor', color='r', linestyle='-', alpha=0.2)

plt.minorticks_on()

plt.grid(True)
plt.show()
print(len(x1))
```
