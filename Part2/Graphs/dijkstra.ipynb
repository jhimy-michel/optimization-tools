```python
import networkx as nx
import tools as helper
from sympy import *
import pandas as pd
from IPython.display import display, HTML


def dijkstra(csvfile, fromnode):
    G, _, _ = helper.get_graph_from_adjacency_matrix(csvfile)
    p = nx.shortest_path(G, source=fromnode, weight='weight')
    df = pd.DataFrame({'target': p.keys(), 'sp': p.values()})
    results = []
    for item in df.get('sp'):
        k = nx.path_weight(G, item, 'weight')
        # print(item)
        results.append([item, k])
    #print(results)
    display(HTML(pd.DataFrame(results, columns=[
            "From - To", "Weight"]).to_html()))


path = "./inputData/citiesData.csv"
dijkstra(path, "Zurich")

helper.print_graph_from_adj_matrix(path)
```
