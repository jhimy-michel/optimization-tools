```python
import networkx as nx
import tools as helper
from sympy import *
import pandas as pd
from IPython.display import display, HTML


def minimumSpanningTree(csvfile):
    G, _, _ = helper.get_graph_from_adjacency_matrix(csvfile)
    T = nx.minimum_spanning_tree(G)
    results = []
    totalweight = 0
    for e in sorted(T.edges(data=True)):
        totalweight += e[2]['weight']
        results.append([e[0], e[1], e[2]['weight']])
    results.append(["----", "TOTAL:", totalweight])
    print(results)
    display(HTML(pd.DataFrame(results, columns=[
            "From", "To", "Weigth"]).to_html()))


path = "./inputData/citiesData.csv"
minimumSpanningTree(path)
helper.print_graph_from_adj_matrix(path)
```
